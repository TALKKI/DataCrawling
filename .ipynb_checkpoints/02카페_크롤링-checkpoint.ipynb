```python
import os
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# ──────────────────────────────────────────────────────────────────────────────
# 1) 환경변수에서 네이버 아이디/비밀번호 읽기 & 검증
# ──────────────────────────────────────────────────────────────────────────────
NAVER_ID = os.getenv("NAVER_ID")
NAVER_PW = os.getenv("NAVER_PW")
if not NAVER_ID or not NAVER_PW:
    raise EnvironmentError(
        "환경변수 NAVER_ID, NAVER_PW를 반드시 설정해주세요.\n"
        "예) export NAVER_ID='아이디'; export NAVER_PW='비밀번호'"
    )

# ──────────────────────────────────────────────────────────────────────────────
# 2) Chrome WebDriver 초기화
# ──────────────────────────────────────────────────────────────────────────────
chrome_opts = Options()
chrome_opts.add_argument("--start-maximized")
# 필요하다면 아래 옵션을 해제/추가하세요.
# chrome_opts.add_argument("--headless")  # 헤드리스 모드
# chrome_opts.add_argument("--disable-gpu")

driver = webdriver.Chrome(options=chrome_opts)

try:
    # ──────────────────────────────────────────────────────────────────────────
    # 3) 네이버 로그인 페이지 이동
    # ──────────────────────────────────────────────────────────────────────────
    driver.get("https://nid.naver.com/nidlogin.login")
    time.sleep(1)  # 페이지 로딩 대기

    # ──────────────────────────────────────────────────────────────────────────
    # 4) 아이디 / 비밀번호 입력 & 로그인 버튼 클릭
    # ──────────────────────────────────────────────────────────────────────────
    driver.find_element(By.ID, "id").send_keys(NAVER_ID)
    driver.find_element(By.ID, "pw").send_keys(NAVER_PW)
    driver.find_element(By.ID, "log.login").click()

    # ──────────────────────────────────────────────────────────────────────────
    # 5) 로그인 처리 대기
    # ──────────────────────────────────────────────────────────────────────────
    time.sleep(2)  # 로그인 처리 시간을 충분히 주십시오.

    # ──────────────────────────────────────────────────────────────────────────
    # 6) 원하시는 카페 페이지로 이동
    #    (예시: 게시글 URL)
    # ──────────────────────────────────────────────────────────────────────────
    cafe_url = (
        "https://cafe.naver.com/f-e/cafes/19997929/articles/90723"
        "?menuid=809&referrerAllArticles=false&inCafeSearch=true&query=시험"
    )
    driver.get(cafe_url)
    time.sleep(2)

    # ──────────────────────────────────────────────────────────────────────────
    # 7) 카페 메인 프레임으로 전환 (네이버 카페는 iframe 내부에 콘텐츠가 있습니다)
    # ──────────────────────────────────────────────────────────────────────────
    driver.switch_to.frame("cafe_main")
    time.sleep(1)

    # ──────────────────────────────────────────────────────────────────────────
    # 8) 댓글/작성자 등 원하는 데이터 크롤링
    #    (예: 첫 댓글 내용 출력)
    # ──────────────────────────────────────────────────────────────────────────
    first_comment = driver.find_element(By.CSS_SELECTOR, ".comment_box .comment_txt")
    print("첫 댓글 내용:", first_comment.text)

    # ──────────────────────────────────────────────────────────────────────────
    # 9) 추가 크롤링 로직을 여기에 작성...
    # ──────────────────────────────────────────────────────────────────────────

finally:
    # ──────────────────────────────────────────────────────────────────────────
    # 10) 브라우저 종료
    # ──────────────────────────────────────────────────────────────────────────
    time.sleep(1)
    driver.quit()
```
