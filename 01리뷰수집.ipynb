```python
from google_play_scraper import Sort, reviews_all
# 앱 리뷰 수집
result = reviews_all(
    'kr.co.lotteins.a1mars',  # 수집할 앱의 패키지 이름을 입력하세요
    sleep_milliseconds=0, 
    lang='ko',            # 리뷰 언어를 한국어로 설정
    country='kr',         # 국가를 한국으로 설정
    sort=Sort.MOST_RELEVANT,  # 가장 관련성 높은 리뷰부터 수집
)
```

```python
import pandas as pd
df = pd.DataFrame(result) # result라는 데이터로 DataFrame을 생성\
df
```

```python
import pandas as pd

# (예시) content_df라는 DataFrame이 이미 존재한다고 가정합니다.
# 여러분의 실제 변수명(content_series 등)을 content_df로 바꿔서 사용하세요.

# 1) 우선, 엑셀에서 허용되지 않는 제어문자 패턴을 하나로 묶어 정규 표현식 작성
#    - [\x00-\x08\x0B-\x0C\x0E-\x1F]  =  허용되지 않는 영역(0x00~0x08, 0x0B~0x0C, 0x0E~0x1F)
illegal_pattern = r'[\x00-\x08\x0B-\x0C\x0E-\x1F]'
content_df = df[['at','content','score']]
# 2) DataFrame 전체 문자열 컬럼에 대해, 위 패턴을 빈 문자열로 치환
content_df = content_df.replace(
    to_replace=illegal_pattern,
    value='',
    regex=True
)

# 3) 그 후에 .to_excel()을 호출하면 더 이상 IllegalCharacterError가 발생하지 않습니다.
output_path = 'content_data_cleaned.xlsx'
content_df.to_excel(output_path, index=False)
print(f"✅ 엑셀 저장 완료: {output_path}")
```
